```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/marketing_campaign.csv', delimiter=';')
count = df['Complain'].sum()
print(count)
```

```python
avg_age_customers = 0
avg_age = (2024 - df['Year_Birth']).mean()
print('Средний возраст людей в датасете = ', avg_age)
mnt_columns = ['MntGoldProds', 'MntFruits', 'MntFishProducts', 'MntMeatProducts', 'MntSweetProducts', 'MntWines']
purchase = df[df[mnt_columns].sum(axis=1) > 0].index
avg_age_customers = (2024 - df.loc[purchase, 'Year_Birth']).mean()
print('Средний возраст покупателей = ', avg_age_customers)
if abs(avg_age - avg_age_customers) == 0:
    print('Средний возраст людей в датасете = средний возраст покупателей')
else:
    print('Средний возраст людей в датасете отличается от среднего возраста покупателей на', abs(avg_age - avg_age_customers))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/marketing_campaign.csv", delimiter=';')
columns = ['Education',
'Marital_Status',
'Income',
'Kidhome',
'Teenhome',
'Recency',
'NumWebPurchases',
'NumStorePurchases']
for i in range(len(columns)):
    for j in range(len(columns)):
        if j > i:
            plt.figure(figsize=(10,5))
            sns.scatterplot(data=df, x=columns[i], y=columns[j])
            plt.show()
            plt.close()
        else:
            continue
for i in range(len(columns)):
    plt.figure(figsize=(10,5))
    sns.scatterplot(data=df, x=2024-df['Year_Birth'], y=columns[i])
    plt.xlabel('Age')
    plt.show()
    plt.close()
```

   3) Реже всего совершают покупки люди с уровнем образования Basic (что логично, тк их прибыль наименьшая)
